```python
# Graph to plot distance and time
distance = list(range(2,11,2))  # [2,4,6,8,10]
time = list(range(1,6))         # [1,2,3,4,5]
# d = 2*t
plt.title("Distance vs Time")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(time,distance,'b*') # r -> red colour,g -> green and b-> blue * -> stars as points, o -> dots as points, + -> + as points
plt.grid()
plt.show()
```

```python
# v = a*t
t = list(range(1,11))
a = list(range(1,len(t)+1))
v = [a[i] * t[i] for i in range(len(t))]

plt.plot(t, v, 'g*')
plt.grid()
plt.show()
a
```

```python
# Numbers vs Cubes
n = list(range(10))
sqn = [i**3 for i in n]
plt.plot(n,sqn)
plt.grid()
plt.show
```

```python
# Voltage - Time
import numpy as np
# t = [range(0.1,2.0,0.1)]  # 'float' object cannot be interpreted as an integer
# t = [i/10 for i in range(1,21)]       (or)
t = np.arange(0.1,2.0,0.1)
v = [1 + np.sin(2*np.pi*i) for i in t]
plt.title("Time vs Voltage Graph")
plt.plot(t,v)
plt.grid()
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.savefig('Images/Voltage - Time.png')  # to save the figure in Images folder
plt.show()
```

```python
# Plot Income of Arkansas from 2005 to 2013
import pandas as pd
df = pd.read_csv('DataFiles/Income.csv')
years = df.columns[2:]
incomeArkansas = df.values[3][2:]
plt.title("Income of Arkansas")
plt.xlabel("Year")
plt.ylabel("Income")
plt.plot(years,incomeArkansas)
plt.show()
```
