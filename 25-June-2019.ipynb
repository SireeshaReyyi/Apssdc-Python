### Marks Analysis Application
    - Generate Marks file for n students
    - Input : Marks text file - Each line in the text file contains marks of one student
    - Generates a report with the following information
        - Class Average
        - % of students passed
        - % of students failed
        - % of students with distinction
        - Highest Mark Frequency
        - Lowest Mark Frequency

```python
from random import randint 

def generateMarks(n,lb,ub,filename):
    with open(filename,'w') as f:
        for i in range(0,n):
            r=randint(lb,ub)
            f.write(str(r)+'\n')
    return
import re
def readFile(filename):
    with open(filename,'r') as f:
        filedata = f.read().split()
    return filedata

def classAverage(filename):
    filedata = readFile(filename)
    tot=0
    for i in filedata:
        tot=tot+int(i)
    return (tot/len(filedata))

def passPercentage(filename):
    filedata = readFile(filename)
    count=0
    for i in filedata:
        i=int(i)
        if i>=35:
            count+=1
    return (count/len(filedata))*100

def failPercentage(filename):
    return (100-passPercentage(filename))
def distinction(filename):
    filedata = readFile(filename)
    count=0
    for i in filedata:
        i=int(i)
        if i>=80:
            count+=1
    return (count/len(filedata))*100
def highestMarkFrequency(filename):
    filedata = readFile(filename)
    count = filedata.count(max(filedata))
    return count
def lowestMarkFrequency(filename):
    filedata = readFile(filename)
    count = filedata.count(min(filedata))
    return count
def classReport():
    print('Class Average :',classAverage(filename))
    print('Pass Percentage :',passPercentage(filename))
    print('Fail Percentage :',failPercentage(filename))
    print('Distinction Percentage :',distinction(filename))
    print('Highest Mark Frequency',highestMarkFrequency(filename))
    print('Lowest Mark Frequency',lowestMarkFrequency(filename))
    return

filename = 'DataFiles/marks.txt'
n=int(input())
# lb=int(input())
# ub=int(input())
generateMarks(n,0,100,filename)
classReport()
readFile(filename)
```
